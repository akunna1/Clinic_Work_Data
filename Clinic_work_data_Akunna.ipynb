```python
import pandas as pd
import numpy as np
import seaborn as sns
print(sns.__version__)
import matplotlib.pyplot as plt

data = pd.read_csv("C:\\Users\\akunna1\\Desktop\\Work_Files\\Mayo_Data_Collection_Akunna_Edited2.csv") 
#The file path would need to be changed when running on a different computer

# Removed unwanted rows (i.e patients 12-21) on excel

#displaying what the first few lines of data looks like
data.head()

# Counting the number of rows and columns
num_rows = data.shape[0]
num_cols = data.shape[1]

# Printing the number of rows and columns
print('Number of rows:', num_rows)
print('Number of columns:', num_cols)
```

```python
# getting the list of column names
column_names = data.columns.tolist()
column_names
```

```python
#calculating the average stress for all participants
stress_mean = np.mean(data['Stress'])
# printing the mean/average of the Stress column
print("The Average Stress =", round(stress_mean,3)) #formating to 3 significant figures
```

```python
#calculating the average sleepiness for all participants
sleep_mean = np.mean(data['Sleep'])
# printing the mean/average of the Sleep column
print("The Average Sleepiness =", round(sleep_mean,3)) #formating to 3 significant figures
```

```python
#calculating the average NASA TLX Score for all participants
NASA_TLX_5_TPmean = np.mean(data['Mean_NASA_TLX_5'])
# printing the mean/average of the Mean_NASA_TLX_5 column
print("The Average NASA TLX Score =", round(NASA_TLX_5_TPmean,3)) #formating to 3 significant figures
```

```python
#calculating the average CSUQ Score for all participants
CSUQ_score_TPmean = np.mean(data['CSUQ_score_avg'])
# printing the mean/average of the CSUQ_score_avg column
print("The Average CSUQ Score =", round(CSUQ_score_TPmean,3)) #formating to 3 significant figures
```

```python
#How the scores vary by Epic Experience
#YRS_EHR - Years of epic/cerner experience
data_select1 = data[['YRS_EHR', '1_Satisfied_Easy', '2_Simple','3_Quick', '4_Comfortable',
                    '5_Easy','6_Productive','7_Error','8_Mistake', '9_Clear','10_Easy_Find',
                     '11_Effective','12_Organization_Clear','13_Pleasant','14_Like',
                    '15_Functions','16_Satisfied','CSUQ_score_avg']]
corr1 = data_select1.corr()
corr1
```

```python
#correlation heatmap for corr1 to show how YRS_EHR (years of experience vary with CSUQ Scores)
plt.figure(figsize=(12,10))
sns.heatmap(data = corr1, annot=True, cmap='rocket_r')
```

```python
#How the scores vary by Epic Experience
#YRS_EHR - Years of epic/cerner experience
data_select2 = data[['YRS_EHR', 'Mental_demand_5', 'Physical_demand_5','Temporal_demand_5', 'Performance_5',
                    'Effort_5','Frustration_5','Mean_NASA_TLX_5']]
corr2 = data_select2.corr()
corr2
```

```python
#correlation heatmap for corr2 to show how YRS_EHR (years of experience vary with MEAN NASA TLX Scores)
plt.figure(figsize=(7,5))
sns.heatmap(data = corr2, annot=True, cmap='mako')
```
